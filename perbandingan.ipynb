```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram untuk semua kolom numerik
df1.hist(figsize=(12,8), bins=15, edgecolor="black")
plt.suptitle("Distribusi Variabel (Histogram)")
plt.show()
```

```python
# Boxplot semua variabel numerik
plt.figure(figsize=(12,6))
sns.boxplot(data=df1)
plt.title("Boxplot Variabel")
plt.xticks(rotation=45)
plt.show()
```

```python
# Heatmap korelasi
plt.figure(figsize=(8,5))
plt.plot(df1["Tahun"], marker="o", color="blue")
plt.title("Tren Tahun Data SULAMPUA")
plt.xlabel("Index")
plt.ylabel("Tahun")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df2, x='Promo(X1)', y='Jual(Y)')
plt.title('Scatter Plot antara Promo dan Penjualan')
plt.xlabel('Promo (X1)')
plt.ylabel('Penjualan (Y)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot dengan garis regresi
plt.figure(figsize=(8, 6))
sns.regplot(data=df2, x='Promo(X1)', y='Jual(Y)', ci=None, line_kws={"color": "red"})
plt.title('Regresi Linear antara Promo dan Penjualan')
plt.xlabel('Promo (X1)')
plt.ylabel('Penjualan (Y)')
plt.grid(True)
plt.show()
```

```python
# Heatmap korelasi
plt.figure(figsize=(6, 4))
sns.heatmap(df2.corr(), annot=True, cmap='coolwarm')
plt.title('Matriks Korelasi')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Buat data Tahun
data = {
    'Tahun': [2013, 2015, 2018, 2021, 2023] * 22  # total 110 data
}
df_tahun = pd.DataFrame(data)

# Visualisasi boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x=df_tahun['Tahun'])
plt.title("Boxplot Distribusi Tahun")
plt.xlabel("Tahun")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
url = "https://raw.githubusercontent.com/Alam-A99/Sulampua/main/DataLN_SULAMPUA.csv"
df1 = pd.read_csv(url, sep=";")

# Ambil nama kolom pertama (Provinsi/Wilayah)
provinsi_col = df1.columns[0]

# Hitung jumlah data per provinsi
provinsi_counts = df1[provinsi_col].value_counts()

# --- Pie Chart ---
plt.figure(figsize=(8,8))
plt.pie(provinsi_counts, labels=provinsi_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title("Distribusi Data per Provinsi")
plt.axis("equal")  # supaya lingkaran bulat
plt.show()



```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df2, x='Promo(X1)', y='Jual(Y)')
plt.title('Scatter Plot antara Promo dan Penjualan')
plt.xlabel('Promo (X1)')
plt.ylabel('Penjualan (Y)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot dengan garis regresi
plt.figure(figsize=(8, 6))
sns.regplot(data=df2, x='Promo(X1)', y='Jual(Y)', ci=None, line_kws={"color": "red"})
plt.title('Regresi Linear antara Promo dan Penjualan')
plt.xlabel('Promo (X1)')
plt.ylabel('Penjualan (Y)')
plt.grid(True)
plt.show()
```

```python
# Heatmap korelasi
plt.figure(figsize=(6, 4))
sns.heatmap(df2.corr(), annot=True, cmap='coolwarm')
plt.title('Matriks Korelasi')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df2 = pd.read_csv("https://raw.githubusercontent.com/Alam-A99/Case-Study/refs/heads/main/Data_Regresi.csv")

# --- Boxplot untuk semua kolom numerik ---
plt.figure(figsize=(10,6))
sns.boxplot(data=df2, orient="v", palette="Set2")
plt.title("Boxplot Data Regresi")
plt.xlabel("Variabel")
plt.ylabel("Nilai")
plt.xticks(rotation=45)
plt.show()
```

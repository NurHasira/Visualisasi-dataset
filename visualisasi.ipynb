```python
# menggabungkan scatter plot dengan tren line
import numpy as np
import matplotlib.pyplot as plt

# Data
x = [3, 6, 8, 10, 6]
y = [5, 7, 9, 12, 15]

# ubah jadi array numpy
x = np.array(x)
y = np.array(y)

# Scatter plot titik-titik yang benar-benar tersebar
plt.scatter(x, y, color='yellow', s=50, label='Data')

# Hitung trend line (regresi linear)
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color='blue', label='Trend Line')

plt.title("Scatter Plot dengan Trend Line")
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.legend()
plt.show()
```

```python
# Mengatur ukuran frame
plt.figure(figsize=(8, 6))
sns.boxplot(data=df,
            x="Promo(X1)", y='Jual(Y)',
            palette="bright",
            linewidth=0.8,
            width=0.5,
            fliersize=3,
            flierprops={"marker": "o"})
plt.legend(bbox_to_anchor=(1, 0.6), loc=2, borderaxespad=1)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 6), sharex=False)
sns.boxplot(data=df, x="Jual(Y)", y='Promo(X1)', width=0.6, ax=axes[0, 0])
sns.boxplot(data=df, x="Karyawan(X2)", y='Jual(Y)', width=0.6, ax=axes[0, 1])
sns.boxplot(data=df, x="Promo(X1)", y='Jual(Y)', width=0.6, ax=axes[1, 0])
sns.boxplot(data=df, x="Promo(X1)", y='Karyawan(X2)', width=0.6, ax=axes[1, 1])

 

fig.tight_layout()
```

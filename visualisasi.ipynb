```python
import matplotlib.pyplot as plt

# Data
x = [2, 5, 10, 12, 10, 20]
y = [4, 9, 12, 30, 17, 22]

# Membuat plot garis
plt.plot(x, y)

# Menampilkan plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [3, 5, 8, 10, 30]
y = [5, 23, 40, 11, 50]

# Membuat scatter plot
plt.scatter(x, y)
# Menambahkan judul dan label
plt.title("Scatter plot")

# Menampilkan plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = ["A", "B", "C", "D", "E"]
y = [100, 70, 50, 20, 40]

# Membuat bar plot
plt.bar(x, y)

# Menambahkan judul dan label
plt.title("Bar Plot")

# Menampilkan plot
plt.show()
```

```python
import matplotlib.pyplot as plt
#Data
data = [20, 90, 40, 12, 40]
# Membuat histogram
plt.hist(data)
#membuat title
plt.title("Histogram")

# Menampilkan plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ["E", "F", "G", "H"]
sizes = [20, 47, 35, 50]

# Membuat pie chart
plt.pie(sizes, labels=labels)

# Menampilkan plot
plt.show()

```
